```python
import pandas as pd
```

```python
df = pd.read_csv("data_wrangling_csv\COVID-19-master\csse_covid_19_data\csse_covid_19_daily_reports\COVID_Data_Today.csv")
```

```python
df = df.set_index(['Country/Region','Province/State'])
df.head(10)
```

```python
myData = df[['Confirmed','Deaths','Recovered']]
myData
```

```python
dataUSA = myData.loc['US']
dataUSA
```

```python
# Get names of indexes for which column Age has value 30
indexNames = dataUSA[dataUSA['Confirmed'] < 1].index
# Delete these row indexes from dataFrame
dataUSA.drop(indexNames , inplace=True)
dataUSA
```

```python
caliData = dataUSA.loc[['California','Diamond Princess','Grand Princess'],:].sum()
caliData
```

```python
caliUSA = pd.DataFrame({'Confirmed':caliData[0],'Deaths':caliData[1],'Recovered':caliData[2]},index = {'California(including Cruise)'})
caliUSA
```

```python
usaData = dataUSA.append(caliUSA)
```

```python
usaData
```

```python
consolidatedData = usaData.drop(['California','Diamond Princess','Grand Princess'])
```

```python
consolidatedData
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt 
import pandas as pd
statesData = consolidatedData.plot.barh(stacked=True,legend=True,title = " US states Coronavirus Data")
statesData.set_xlabel("Number of cases")
statesData.set_ylabel("US States")

```
